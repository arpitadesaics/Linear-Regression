```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel(r"C:\Users\DELL\Desktop\Data Science\Files\boston housing.xls")
df.head()
```

```python
x = df[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PT','B','LSTAT']]
y = df['MV']
model = LinearRegression()
model.fit(x, y)
```

```python
p_value = model.predict(x)
p_value
```

```python
corr = df.corr()
corr
```

```python
r_sq = model.score(x, y)
r_sq
```

```python
plt.plot(corr)
```
